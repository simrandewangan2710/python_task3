Qu. Prove that the string is Palindrome or not?
str1 = 'a,b,c,d,c,b,a'
##### 2 string
##### 'abcdcba' == 'abcdcba'
abcdcba' == 'abcdcba'

```python
str1 = 'a,b,c,d,c,b,a'
Newstr = str1.replace(',','')
print('Newstr :',Newstr)
Revstr = Newstr[::-1]
print('Revstr:',Revstr)
print(f"Is Palindrome ?: {Newstr == Revstr}")
```
